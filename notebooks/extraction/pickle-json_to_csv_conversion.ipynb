```python
import pandas as pd
import pickle
import json
import datetime as dt
import geopy
import unidecode
```

```python
def gen_date_ranges(start_date, end_date):
    dates = []
    if start_date.month == end_date.month:
        dates.append([start_date.strftime("%Y-%m-%d"), end_date.strftime("%Y-%m-%d")])
    else:
        data_ranges = pd.date_range(start= start_date, end = end_date, freq="M")
        for date_end in data_ranges:
            date_start = dt.date(date_end.year,date_end.month,1).strftime("%Y-%m-%d")
            date_end = date_end.strftime("%Y-%m-%d")
            dates.append([date_start, date_end])
        dates[-1][1] = end_date.strftime("%Y-%m-%d")
    return dates
```

```python
months = {'Jan':'01', 'Feb':'02', 'Mar':'03', 'Apr':'04', 'May':'05', 'Jun':'06', 'Jul':'07', 'Aug':'08', 'Sep':'09', 'Oct':'10', 'Nov':'11', 'Dec':'12'}
def obt_date(datetime = str):
    year = datetime.split(" ")[-1]
    month = months[datetime.split(" ")[1]]
    day = datetime.split(" ")[2]
    return(year+"-"+month+"-"+day)
def reset_index_df(loaded):
    loaded = loaded.reset_index(drop = True)
    for i in range(len(loaded)):
        if loaded['Id'][i] == 0:
            loaded = loaded.drop(loaded.index[i])
    loaded = loaded.reset_index(drop = True)    
    return(loaded)
```

```python
def add_to_dict(new_df):
    texts_dict = []
    for i in range(len(new_df)):
        id_i = new_df['Id'][i]
        text_i = new_df['text'][i]
        date_i = obt_date(new_df['json'][i]['created_at'])
        try:
            country_i = new_df['location'][i].address.split(",")[-1]
            country_i = country_i.strip()
        except:
            country_i = ""
        texts_dict.append({'Id':id_i, 'text':text_i, 'date':date_i, 'country':country_i})                   
        
    return(texts_dict)
```

```python
texts = []
#Range is 2015-01 to 2021-03
dates = ['2015-01', '2015-02', '2015-03', '2015-04', '2015-05', '2015-06', '2015-07', '2015-08', '2015-09', '2015-10', '2015-11', '2015-12'] #data range from 2015-01 to 2015-12
dates = dates + ['2016-01', '2016-02', '2016-03', '2016-04', '2016-05', '2016-06', '2016-07', '2016-08', '2016-09', '2016-10', '2016-11', '2016-12'] #data range from 2016-01 to 2016-12
dates = dates + ['2017-01', '2017-02', '2017-03', '2017-04', '2017-05', '2017-06', '2017-07', '2017-08', '2017-09', '2017-10', '2017-11', '2017-12'] #data range from 2017-01 to 2017-12
dates = dates + ['2018-01', '2018-02', '2018-03', '2018-04', '2018-05', '2018-06', '2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12'] #data range from 2018-01 to 2018-12
dates = dates + ['2019-01', '2019-02', '2019-03', '2019-04', '2019-05', '2019-06', '2019-07', '2019-08', '2019-09', '2019-10', '2019-11', '2019-12'] #data range from 2019-01 to 2019-12
dates = dates + ['2020-01', '2020-02', '2020-03', '2020-04', '2020-05', '2020-06', '2020-07', '2020-08', '2020-09', '2020-10', '2020-11', '2020-12'] #data range from 2020-01 to 2020-12
dates = dates + ['2021-01', '2021-02', '2021-03'] #data range from 2021-01 to 2021-03

for date in dates:
        try:
            loaded = pickle.load(open("./../data/monthly_pickles/df"+date+".pickle","rb"))
            loaded = reset_index_df(loaded)
            texts = texts + add_to_dict(loaded)
            print(date+" loaded")
        except Exception as r:
            print(date+" not loaded "+str(r))
```

```python
texts = pd.concat([previous_texts,pd.DataFrame.from_dict(texts)], ignore_index = True)
texts['country'] = texts['country'].fillna('no_country').str.strip()
texts['country'] = texts['country'].apply(lambda x: unidecode.unidecode(x))
texts.to_csv('../data/texts/texts_from_pickles.csv', index = False)
```
